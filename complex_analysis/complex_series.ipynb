# Plot $z^n$ serie

An infinite series of complex numbers $\sum_{0}^{\infty}{z^n}$:
* |z|<1, $\sum_{0}^{\infty}{z^n}$ converge to $S=\frac{1}{1-z}$
* |z|>1, $\sum_{0}^{\infty}{z^n}$ diverge

We will plot its partial sums $S_N = \sum_{0}^{N}{z^n}$, that is to say, plot each point {$S_0, S_1, S_2, S_3, ..., S_N$}


```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import *
def module(z):
    return np.sqrt(z.real*z.real+z.imag*z.imag)
def plot_z(z,num_iters=20):
    
    sum=0
    sums=[]
    for i in range(num_iters):
        sum +=power(z,i)
        sums.append(sum)

    X=[i.real for i in sums]
    Y=[i.imag for i in sums]

    m = max(max([abs(i)for i in X]),max([abs(i) for i in Y]))

    plt.figure(figsize=(8, 8))
    plt.plot([-1.1*m,1.1*m],[0,0],'k',linewidth=0.5)  # x axis
    plt.plot([0,0],[-1.1*m,1.1*m],'k',linewidth=0.5)  # y axis
    plt.plot(X, Y, 'ob-')
    plt.show()
    mod=module(z)
    if mod<1:
        print("|z|=",mod, "|z|<1, converge")
    else:
        print("|z|=",mod, "|z|>=1, diverge")
    #plt.savefig(savename)

plot_z(0.5+0.5j)
plot_z(0.4+1.1j)
```
