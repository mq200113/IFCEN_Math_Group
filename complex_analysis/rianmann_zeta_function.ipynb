# Rianmann Zata function

```python
def zeta(s):
  ret=0
  for n in range(1,50):
    ret+=power(n,-s)
  return ret
```

```python

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import *

def plot_zeta(s):
    sum=0
    sums=[]
    for n in range(1,30):
        sum +=power(n,-s) 
        sums.append(sum)

    X=[i.real for i in sums]
    Y=[i.imag for i in sums]

    m = max(max([abs(i)for i in X]),max([abs(i) for i in Y]))
    plt.figure(figsize=(8, 8))
    plt.plot([-1.1*m,1.1*m],[0,0],'k',linewidth=0.5)  # x axis
    plt.plot([0,0],[-1.1*m,1.1*m],'k',linewidth=0.5)  # y axis
    plt.plot(X, Y, 'ob-')
    plt.plot(s.real,s.imag,'or')
    plt.show()
    #plt.savefig(savename, bbox_inches='tight')
```

```python
plot_zeta(1.5+1.8j)
```

```python
plot_zeta(1.5-1j)
```
