# Archimedes spiral

Archimedes spiral is the trajectory of a point moving with a constant speed along a line that rotates with constant angular velocity.

$$ r=a\theta$$

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def spiral(a=3):
    x = []
    y = []
    thetas=[i*np.pi*2/100 for i in range(100)]
    for i in range(100):
        theta = thetas[i]
        r = a * thetas[i]
        xi = r * np.cos(theta)
        yi = r * np.sin(theta)
        x.append(xi)
        y.append(yi)
    plt.plot(x,y,"r*")
    plt.show()
spiral()
```

# Amazing!

“The surface of the first round of the spiral is equal to one third of the surface of the circle of while the radius is the length traveled by the point on the straight line during the first round"

螺线第一圈与初始线所围的面积等于第一个圆的三分之一

```python

def spiral_surface(a=3):
    x = []
    y = []
    N=1000
    thetas=[i*np.pi*2/N for i in range(N)]
    for i in range(N):
        theta = thetas[i]
        r = a * thetas[i]
        xi = r * np.cos(theta)
        yi = r * np.sin(theta)
        x.append(xi)
        y.append(yi)
    plt.figure(figsize=(8, 8))
    # plot spiral
    plt.plot(x,y,"r-")

    # plot surface between spiral and x-axis
    y0=[0 for i in range(N)]
    plt.fill_between(x,y,y0,facecolor='red',alpha=0.3)

    # plot circle
    radius= x[-1]
    xx = [radius*np.cos(i) for i in thetas]
    yy = [radius*np.sin(i) for i in thetas]
    plt.plot(xx,yy)
    plt.show()

spiral_surface()
```
